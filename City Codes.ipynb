```python
from json import loads
import pandas as pd
from requests import get
```

```python
url = 'https://safar724.com/route/getcities'
cities = loads(get(url).content)
cities = pd.json_normalize(cities)
cities
```

```python
# duplicate cities

cities[cities['PersianName'].duplicated()]
```
